## if문

### if 문의 기본 구조
#### if 조건문: 조건문이 참이면 다음 문장들을 수행한다.
>수행할 문장1

>수행할 문장2

>...

#### else: 조건문이 참이 아니면 다음 문장들을 수행한다.
>수행할 문장A

>수행할 문장B

>...

### 들여쓰기 (indentation)
들여쓰기 정렬을 제대로 하지 않으면 오류가 발생한다.

space를 사용할 수도 있고, tab을 사용할 수도 있다.

```python
# Code 1
# 다음은 무엇을 하는 코드인가?
a = 2
b = 2
if a == b:
    print(a)
else:
    print("다르다")
```

Code 1에서 보듯이 if 다음에는 bool type의 결과가 와야한다.

즉, True인지 False인지에 따라 True이면 if 문 다음 줄에 오는 문장들을 수행하고 그렇지 않으면 else 아래 줄에 있는 문장들을 수행하게 된다.

### 비교 연산자
비교연산|True인 경우
--|--
x<y|x가 y보다 작을 때
x>y|x가 y보다 클 때
x==y|x와 y의 값이 같을 때
x!=y|x와 y의 값이 다를 때
x>=y|x가 y보다 크거나 같을 때
x<=y|x가 y보다 작거나 같을 때


### 다양한 조건을 판단하고 싶을 때는 elif를 사용한다.
예) "주머니에 돈이 있으면 택시를 타고, 주머니에 돈은 없지만 카드가 있으면 택시를 타고, 돈도 없고 카드도 없으면 걸어 가라"

```python
# 이중 if문을 사용하여 if와 else만으로 구현할 수 있다. 
pocket = ['paper', 'handphone']
card = True
if 'money' in pocket:
    print("택시를 타고가라")
else:
    if card:
        print("택시를 타고가라")
    else:
        print("걸어가라")
```

```python
# elif를 사용하면 좀더 깔끔하게 구현할 수 있다.
pocket = ['paper', 'cellphone']
card = True
if 'money' in pocket:
    print("택시를 타고가라")
elif card: 
    print("택시를 타고가라")
else:
    print("걸어가라")
```

## while문
while 다음에 오는 조건문이 True인 동안에는 그 아래에 있는 문장들을 반복적으로 수행한다.

while 조건문: 조건문이 참이면 다음 문장들을 순차적으로 수행한 후 다시 조건문을 검사하고 참이면 수행하는 과정을 반복적으로 진행한다.

수행할 문장1

수행할 문장2

...

### while문 예제

```python
# Code 7
treeHit = 0
while treeHit < 10:
    treeHit = treeHit +1 # treeHit += 1  으로 표현할 수도 있다.
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

## for문


### for문의 기본 구조
for 변수 in 리스트(또는 튜플, 문자열):
>수행할 문장1

>수행할 문장2

>    ...

리스트나 튜플, 문자열의 첫 번째 요소부터 마지막 요소까지 차례로 변수에 대입되어 "수행할 문장1", "수행할 문장2" 등이 수행된다.

### for문 기본예제

```python
# Code 11
# 리스트를 사용하는 예
test_list = ['one', 'two', 'three'] 
for i in test_list: # 리스트에 있는 요소들을 순차적으로 끄집어내서 변수 i에 할당한다.
    print(i)
```

```python
# Code 12
a = [(1,2), (3,4), (5,6)] # 튜플들을 요소로 갖는 리스트
for first, last in a: # 리스트에 있는 튜플들을 하나씩 꺼낸 후, 튜플의 요소들을 first와 last에 할당한다.
    print(first + last)
```

### for문과 range함수
for문은 숫자 리스트를 자동으로 만들어 주는 range라는 함수와 함께 사용되는 경우가 많다.

다음은 range 함수의 간단한 사용법이다.

```python
a = range(10)
print(a) # 리스트가 아니라 range 객체가 출력된다. 0부터 10미만의 숫자를 포함.
```

```python
# Code 15
a = range(10)
print(a) # 리스트가 아니라 range 객체가 출력된다. 0부터 10미만의 숫자를 포함.

for i in range(10):
    print(i)
```

```python
# Code 16
# 무엇을 하는 코드인가?
sum = 0
for i in range(1, 11): # 1부터 10까지
    sum = sum + i
print(sum)
```

```python
# Code 18
for i in range(2,10):        # ①번 for문
    for j in range(1, 10):   # ②번 for문
        print(i*j, end=" ") # end=" "을 추가하여 해당 결과값을 출력할 때
                         # 다음줄로 넘기지 않고 공백 하나를 붙인 후 계속 이어서 출력하게 한다
    print('') # 이게 없으면 계속 오른쪽에 이어서 출력된다.
```

**for문이나 if문은 들여쓰기(indentation)를 잘못하면 의도하지 않은 결과를 초래할 수 있으므로 주의해야 한다.**

아래 두 코드 Code 24, Code 25은 들여쓰기의 차이만 있는데 다른 결과가 출력됨을 알 수 있다.

```python
# Code 24
for i in range(2):
    for j in range(3):
        print(j)
    print(i)
```

```python
# Code 25
for i in range(2):
    for j in range(3):
        print(j)
        print(i)
```
