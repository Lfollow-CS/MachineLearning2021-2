네 가지 속성(feature):

꽃받침 길이(sepal length), 꽃받침 너비(sepal width), 꽃잎 길이(petal length), 꽃잎 너비(petal width)

150개 데이터

```python
from sklearn.datasets import load_iris 

iris = load_iris() 
iris.data[:5]     # 최초 5개 데이터의 값을 출력
```

```python
import pandas as pd

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = pd.Series(iris.target)
iris_df.head()
```

### 붓꽃 데이터로 학습하기

```python
X = iris_df.iloc[:, :4]
y = iris_df.iloc[:, -1]
y
```

```python
# 또는 아래와 같이
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics 
```

### train_test_split()

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html


```python
k=5
X_train,X_test,y_train,y_test =train_test_split(X, y) # test_size=0.25
knn = KNeighborsClassifier(k) 
knn.fit(X_train, y_train) 
```

```python
y_pred = knn.predict(X_test) 
accuracy_score = metrics.accuracy_score(y_test, y_pred) 
print(knn.score(X_test, y_test))
print(accuracy_score)
```

### Model evaluation
https://scikit-learn.org/stable/modules/model_evaluation.html

```python
from sklearn.metrics import confusion_matrix
print(y_test)
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat.T)
conf_mat
```

```python
metrics.recall_score(y_test, y_pred, average=None)
```

```python
metrics.precision_score(y_test, y_pred, average=None)
```

```python
metrics.f1_score(y_test, y_pred, average=None)
```
