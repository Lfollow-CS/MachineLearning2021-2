## 튜플 자료형 (Tuple)
몇 가지 점을 제외하곤 리스트와 거의 비슷하며 리스트와 다른 점은 다음과 같다.
- 리스트는 [과 ]으로 둘러싸지만 튜플은 (과 )으로 둘러싼다.
- 리스트는 요소의 추가, 삭제, 수정이 가능하지만 튜플은 그 값을 바꿀 수 없다.
- 리스트를 훨씬 더 많이 사용하게 되지만, 값을 바뀌지 않아야 하는 경우에는 튜플을 사용할 수 있다.

```python
# Code 28
t1 = ()
t2 = (1,)
t3 = (1, 2, 3)
t4 = 1, 2, 3 # 튜플은 괄호를 생략해도 된다.
t5 = ('a', 'b', ('ab', 'cd'))
```

### 튜플 연산
값을 변경하는 것을 제외하고는 튜플은 리스트와 같다.

```python
# Code 29
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
print( t1 + t2 )
```

```python
# Code 30
t2 = (3, 4)
print(t2[1])
print( t2 * 3 )
```

## 딕셔너리 자료형 (Dictionary)
사전처럼 하나의 요소가 다른 요소와 쌍을 이룬다.

예) 학교: school, 학생: student

앞의 요소를 Key라고 하고 Key에 대응되는 요소를 Value라 한다.

Key는 변경할 수 없지만, Value는 변경할 수 있다.

Key와 Value는 항상 '짝'을 지어 존재하며, 두 값 사이에는 : 기호로 구분한다.

```python
# Code 31
dic = {'name':'John', 'age':28, 'nationality':'USA'}
print(dic)
```

위에서 'name', 'age', 'nationality'는 Key이고, 각 Key에 해당하는 Value들은 'John', 28, 'USA'이다.

```python
# Code 32
a={1:'one'}
b={1:2}
c={'friends':['John','Bob','Ted']}
d={'friends':('John','Bob','Ted')}
```

위에서처럼 Value는 숫자, 문자열, 리스트, 튜플 등 어떤 형태도 가능하다.

아래처럼 딕셔너리의 Value로 또 다른 딕셔너리를 사용할 수도 있다. (이중 딕셔너리)

```python
# Code 34
dic = {'friends':{'age':25, 'hair':'short'}}
print(dic['friends'])
print(dic['friends']['age'])
```

### 딕셔너리 사용하기
리스트와 달리 요소를 추출할 때 인덱스나 슬라이싱을 사용하지 않고 Key를 이용하여 Value를 추출한다.

```python
# Code 37
grade = {'John': 'A+', 'Tom': 'A-'}
print(grade['John'])
print(grade['Tom'])
```

## Bool 자료형
참(True)과 거짓(False)을 나타내는 자료형이다.

불 자료형은 다음의 2가지 값만을 가질 수 있다.

- True - 참
- False - 거짓

항상 첫 문자를 대문자로 써야 한다. (첫문자만)
true 또는 TRUE로 쓸 수 없다.

```python
# Code 45
a = True
print(a)
b = False
print(b)
```

```python
# Code 46
print(2>1) # 2는 1보다 크다라는 명제는 참이므로 True가 된다.
b=1>2
print(b)
```

```python
# Code 47
print( 'John' in grade)
print( 'Steve' in grade )
```

```python
# Code 48
a=1
b=2
print( a == b )
```

### 자료형의 참과 거짓
어떤 자료형이든 변수가 비어 있거나 0이 들어 있으면 Bool type으로는 False가 된다. 반대로 0이 아닌 어떤 값이 들어 있다면 Bool type으로는 True가 된다.

```python
# Code 49
print(bool(0))
print(bool(-1))
```

```python
# Code 50
print(bool('abc'))
```

```python
# Code 51
# 비어있는 경우
a=""
print(bool(a))
a=[]
print(bool(a))
a={}
print(bool(a))
```
