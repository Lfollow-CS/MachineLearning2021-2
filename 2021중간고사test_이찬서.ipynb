####[2] [1]에서 로딩한 iris data에서 label 2를 label 1로 만드시오.

```
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
iris_new = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_new
```

```
iris_new = pd.DataFrame(iris.data, columns=iris.feature_names) 
iris_new = iris_new[[iris.feature_names[0], iris.feature_names[2],iris.feature_names[1],iris.feature_names[3]]]
iris_new
```

####[3] Support vector machine (SVM) 분류기를 이용하여 모델을 학습하고 정확도를 평가하시오. Label 0과 label 1 데이터의 각각 80%를 학습데이터로 이용하고 나머지 20%를 테스트데이터로 이용하시오. SVM 분류 모델을 만들기 위해 아래 사이트에 있는 클래스를 이용하시오. Kernel로는 linear와 rbf를 사용하여 성능을 비교해보시오. C=10으로 하고, rbf 커널을 위해서는 gamma를 0.1로 두시오.
https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC

*본 문제와 같이 수업시간(실습)에 다뤄보지 않은 내용을 sklearn 사이트(또는 인터넷 검색)를 참조하여 코드를 작성하는 문제가 출제될 수도 있으니 대비하세요.



```
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
import sklearn.metrics as metric
```

```
X = iris_new.iloc[:,:4]
y = iris.target
X_train,X_test,y_train,y_test =train_test_split(X, y, train_size=0.8)
X.shape
```

```
linear = SVC(kernel='linear', C=10)
linear.fit(X_train, y_train)
```

```
rbf = SVC(kernel='rbf', C=10, gamma=0.1)
rbf.fit(X_train, y_train)
```

```
Y_linear = linear.predict(X_test)
```

```
Y_rbf = rbf.predict(X_test)
```

```
print('Accuracy linear=', accuracy_score(y_test, Y_linear))
print('Accuracy rbf=', accuracy_score(y_test, Y_rbf))
```

```
linear.score(X_test, y_test)
```

```
rbf.score(X_test, y_test)
```
