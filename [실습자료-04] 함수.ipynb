## 함수의 정의
def를 이용하여 함수를 정의한다.

```python
# Code 1
# a와 b가 입력으로 들어왔을 때 두 매개변수의 합을 리턴하는 함수를 정의함
def add(a,b):
    return a+b
```

```python
# Code 2
# add 함수를 사용하여 두 숫자를 더해보자
a = 4
b = 3
c = add(a, b) # add함수에서 리턴하는 값을 c에 저장한다.
print(c)
```

```python
# Code 3
# 함수가 아무값도 리턴하지 않을 수도 있다.
def add_print(a,b):
    print(a+b)
  
add_print(4,3)
```

```python
# Code 4
# 함수에 아무 입력이 없을 수도 있다.
def print_Hello():
    print("Hello!")

print_Hello()
```

함수가 정의된 부분이 먼저 실행되고 나서야 그 함수를 사용할 수 있다.

아래와 같이 하면 오류 발생!

```python
# Code 5
print_Hi()

def print_Hi():
    print("Hi!")
```

## 매개변수를 지정하여 함수 호출하기
매개변수를 지정하면 매개변수의 순서에 신경쓰지 않아도 된다.

```python
# Code 6
def subtract(a, b):
    return a-b

# 7 - 3을 계산하고 싶을 때
print(subtract(7, 3))
print(subtract(a=7, b=3))
print(subtract(b=3, a=7))
```

## 함수의 결과값은 언제나 하나이다

```python
# Code 11
def add_and_mul(a,b): 
    return a+b, a*b
```

위의 코드는 a+b와 a\*b 두 개의 값을 출력으로 리턴하는 것처럼 보이지만 사실은 괄호가 생략된 하나의 튜플을 리턴하는 것이다.
아래의 코드를 실행해서 result를 확인해보자.

```python
# Code 12
result = add_and_mul(3,4)
print(result)
```

다음의 코드처럼 튜플를 리턴하는 함수의 출력을 서로 다른 두 변수로 받을 수도 있다.

```python
# Code 13
result1, result2 = add_and_mul(3, 4)
print(result1)
print(result2)
```

## 매개변수에 default 값 설정하기
해당 매개변수를 입력하지 않으면 default값이 사용된다.

```python
# Code 15
def say_myself(name, age, man=True): 
    print("나의 이름은 %s 입니다." % name) 
    print("나이는 %d살입니다." % age) 
    if man: 
        print("남자입니다.")
    else: 
        print("여자입니다.")
        
say_myself("홍길동", 20) # man에 해당하는 입력변수를 넣지 않아서 default값이 사용된다.
say_myself("홍길동", 20, True)
say_myself("심청", 18, False)
```
